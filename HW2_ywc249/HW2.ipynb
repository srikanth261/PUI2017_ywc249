```python
#apikey = os.getenv("MTAAPIKEY")
#apikey = sys.argv[0]
apikey = "6f7345b7-da51-4ff3-8bde-fca90df6072a" ### DONT SHOW ###

#bus = sys.argv[1]
bus = "B8"

#url = "http://api.openweathermap.org/data/2.5/forecast?q=" + "NewYork&mode=Json&units=metric&cnt=7&APPID=" + os.getenv("OPENWEATHKEY")
url = "http://bustime.mta.info/api/siri/vehicle-monitoring.json?version=1&key=%s&VehicleMonitoringDetailLevel=calls&LineRef=%s"%(apikey, bus)

#use the json.loads method to obtain a dictionary representation of the responose string 
response = urllib.urlopen(url)
data = response.read().decode("utf-8") ### DECODE IS NOT NECESSARY FOR PYTHON 3 ###
data = json.loads(data)
```

```python
data
```

```python
try:
    data2 = data['Siri']['ServiceDelivery']['VehicleMonitoringDelivery'][0]['VehicleActivity']
except:
    error = data['Siri']['ServiceDelivery']['VehicleMonitoringDelivery'][0]['ErrorCondition']
    print(error['Description'])
    sys.exit()
```

```python
data2
```

```python
lat = []
lng = []
vref = []
print("Bus Line: " + bus)
print("Number of Active Buses: " + str(len(data2)))
i = 1
for v in data2:
    global lat, lng, vref
    line = v['MonitoredVehicleJourney']['LineRef']
    lt = v['MonitoredVehicleJourney']['VehicleLocation']['Latitude']
    lg = v['MonitoredVehicleJourney']['VehicleLocation']['Longitude']
    vrf = v['MonitoredVehicleJourney']['VehicleRef']
    vrf = vrf.split("_")[1]
    print("Bus %s (#%s) is at latitude %s and longitude %s"%(i, vrf, lt, lg))
    lat.append(lt)
    lng.append(lng)
    vref.append(vrf)
    i += 1
```

```python
data2[0]['MonitoredVehicleJourney'].keys()
```

```python
status = []
stop = []
for v in data2:
    global status, stop
    if 'OnwardCall' in v['MonitoredVehicleJourney']['OnwardCalls']:
        i = v['MonitoredVehicleJourney']['OnwardCalls']['OnwardCall'][0]
        stt = i['Extensions']['Distances']['PresentableDistance']
        stp = i['StopPointName']
    else:
        stt = 'N/A'
        stp = 'N/A'
    status.append(stt)
    stop.append(stp)
```

```python
d = {'Bus #': vref, 'Latitude': lat, 'Longitude': long, 'Stop Name': stop, 'Stop Status': status}
df = pd.DataFrame(data = d)
df.index += 1
df
```

```python
with open('%s.csv'%(bus), 'w') as f:
    df.to_csv('%s.csv'%(bus))
"""
file = sys.argv[3]
with open('%s'%(file), 'w') as f:
    df.to_csv('%s'%(file))
"""
```
